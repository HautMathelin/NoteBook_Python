## 1. Gratuit et Open Source
Python est un langage de programmation gratuit, open source, libre et multi-plateforme.


### a. Open source
Python est un langage *Open Source* car son code source (texte lisible et compréhensible par l'homme qui a vocation à être traduit en code binaire directement utilisable par la machine) est disponible.


### b. Libre
Sa <span style="color: red; font-weight: bold;">licence d'utilisation</span> (contrat fixant les droits et les devoirs de l'utilisateur) respecte l'esprit du <span style="color: red; font-weight: bold;">logiciel libre</span>
* possibilité de *l'utiliser*
* possibilité *d'étudier le code source*
* possibilité de *redistribution*,
* possibilité de *modification*


### c. Multi-plateforme
Un <span style="color: red; font-weight: bold;">système d'exploitation</span> est le logiciel principal d'un ordinateur. 
* Il s'active au démarrage de l'ordinateur. 
* Il sert d'intermédiaire entre les logiciels et le matériel informatique (écran, souris, clavier, disques durs, cartes réseaux, mémoire, ...)
* Tous les autres programmes utilisent ses services


Python tourne sur de multiples systèmes d'exploitation: Windows, Linus, Mac OS, Android, iOS, ...


## 2. De quoi a-t-on besoin pour faire du Python ?

### a. Interpéteur Python
Il est indispensable de disposer d'un *interpréteur Python* qui traduit et exécute sur une machine les instructions écrites par le programmeur en langage Python.

Il existe de nombreux interpéteurs Python. 


### b. Editeur 
L'éditeur est un mini-traitement de texte qui permet de taper le code Python.


### c. EduPython
<img src="edupython.jpg" alt="Icone d'EduPython" width="196px">

*EduPython* est un un EDI (en anglais **E**nvironement **I**ntégré de **D**éveloppement) installé sur les postes du lycée pour faire du Python. Il permet:
* d'éditer du code avec le logiciel *PyScripter* (taper des instructions, les sauvegarder, ...)
* d'interpréter (et exécuter) un programme écrit dans le langage Python avec l'interpréteur *Conda*
* et accessoirement de lancer le serveur pour travailler avec des notebook

## 3. Utiliser EduPython
Il n'y a rien à mémoriser dans ce paragraphe: juste comprendre ce que vous faites !

### a. Uniquement en mode console 

Le *mode console* fonctionne en gros comme une calculatrice: chaque ligne tapée est interprétée au fur et à mesure qu'elle est entrée.

1. Activer la fenêtre *PyScripter* et cliquer dans le cadre de la **console** (en bas par défaut)

2. Taper successivement (faire *Entrée* entre chaque ligne) en observant bien à chaque fois ce qui apparait à l'écran<br>
a = 20<br>
b = a + 100<br>
10*b

<img src="console.jpg" alt="mode console">
Le résultat de $10\times b$ est écrit à l'écran


### b. Uniquement en mode script

Avec le *mode script* (on écrit d'abord tout le code puis on l'interpète), vous commencez par écrire tout votre code dans un éditeur de texte puis vous lancez l'interpréteur qui va exécuter à la suite toutes les lignes de votre code.

1. Activer la fenêtre *PyScripter* et cliquer dans le cadre de **l'editeur** (en haut par défaut)


2. Taper<br>
a = 20<br>
b = a + 100<br>
10*b


3. Cliquer sur la flèche verte dans la barre d'outil (en haut par défaut). Cela transmet à l'interpéteur ce que vous venez de taper dans l'éditeur qui l'exécute.
![mode script](script.jpg "Bouton pour exécuter le code")


4. Pourquoi on ne voit rien ? Parce qu'on n'a pas demandé d'afficher quelque chose à l'écran.<br>
Dans votre code, à la fin, ajouter la ligne <span style="text-align:left; background: lightyellow ; border: 1px solid #999; margin: 5px 0 ; padding: 2px ; width:20%">print(10 * b)</span>: 
cette instruction affiche à l'écran le résultat de $10\times b$


5. Relancer l'éxécution. Cette fois le résultat va s'afficher.


### c. Ecrire le code en mode script et l'exécuter en mode console

1. Dans le cadre de **l'editeur** (en haut par défaut), taper:<br>
c = 20<br>
d = c + 100<br>


2. Cliquer sur la flèche verte dans la barre d'outil (en haut par défaut). Cela transmet à l'interpéteur ce que vous venez de taper dans l'éditeur qui l'exécute.<br>
Maintenant, la variable *c* contient 20 et *d* contient 120 (le résultat de $c + 100$)


3. Dans le cadre de la **console**, taper <span style="text-align:left; background: lightyellow ; border: 1px solid #999; margin: 10px 0 ; padding: 10px ; width:20%">10 * d</span><br>
Le résultat 120 va s'afficher.
